### Histogram example

```python
# you can chooss the style
plt.style.use(style)
# customize the size of the plot
fig, ax = plt.subplots(figsize=(20, 10))
# count number of unique values in the wanted column
nbr_of_cardinality = int(africa['SQACADEM'].nunique())
# plot the histogram
fig = africa["SQACADEM"].hist(
                        method="count",
                        max_cardinality=nbr_of_cardinality, # number of unique values of the column
                                                      # when you use the nbr of unique values the bar get sorted auto
                        ax = ax,
                        color='#66b3ff',
                        log=False

                        )
```

### Pie chart example

```python
pie_kwds = {
    # ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0','#ffb3e6'],
    "colors": None,
    "pctdistance": 0.2,
    "autopct": '%1.1f%%',
    "labeldistance": 0.6,
    "startangle": 45,
    "radius": 1.3,
}
```

```python
# you can chooss the style
plt.style.use(style)
# count number of unique values in the wanted column
nbr_of_cardinality = int(africa["PLIGHT"].nunique())
# customize the size of the plot
# fig, ax1 = plt.subplots(figsize=(20, 10))
africa["PLIGHT"].pie(
    method="count",
    max_cardinality=nbr_of_cardinality,
    pie_type="auto",  # auto,donut, rose
    **pie_kwds
)
```

### Bar plot example

```python
plt.style.use(style)

bar_kwds = {"align": 'center'}
# customize the size of the plot
fig, ax = plt.subplots(figsize=(30, 20))
africa.bar(columns=['country', 'XSEX'],
           method="count",
           of="",
           max_cardinality=(int(africa['country'].nunique()),  # choose the cardinality of the column
                            int(africa['XSEX'].nunique())),
           hist_type="auto", # stacked, pyramid, fully_stacked
           ax=ax,
           **bar_kwds
           )
```

### Boxplot chart example

```python
numcols = ["XNUMYRS", "ZMALEVP", "SINS2006", "numstu"]
```

```python
# Style of the plot
plt.style.use(style)
# set the axis
fig, ax = plt.subplots(figsize=(30, 10))
# boxplot 
africa.boxplot(columns = numcols, 
                ax=ax,
                   )
```

### Scatter plot example

```python
# Style of the plot
plt.style.use(style)
# set the axis
fig, ax = plt.subplots(figsize=(30, 10))
# number of cardinality
nbr_cardinality = int(africa['country'].nunique())
# set the bounderies, it will not work if the number of columns more than 2
xmin = -10
xmax = 20
ymin = -10
ymax = 20
# Scatter plot
africa.scatter(numcols,
               catcol="country",
               max_cardinality=nbr_cardinality,
               cat_priority=[],
               with_others=False, # True: aggregate the small representavive set, False will not aggregate them and not plot them
               max_nb_points=20000,
               dimensions=None,
               bbox=[xmin, xmax, ymin, ymax], # the bounderies of the plot
               ax=ax,)
```
